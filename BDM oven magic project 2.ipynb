```python
import pandas as pd
import re
from mlxtend.frequent_patterns import apriori


# Function to extract clean item names
def extract_clean_items(row):
    clean_items = []
    for item in row:
        if isinstance(item, str) and item != 'nan':
            item_name = re.search(r'([a-zA-Z\s]+)', item)
            if item_name:
                clean_items.append(item_name.group(1).strip())
    return clean_items

# Apply the extract_clean_items function to create clean item columns
data['clean_items'] = data['item_purchased'].apply(extract_clean_items)

# Display clean items
print(data['clean_items'])

# Combine all clean item columns into a single list
transaction_list = data['clean_items'].values.tolist()

# Convert to binary encoded format and then to boolean
binary_encoded = pd.DataFrame(transaction_list).applymap(lambda x: 1 if x else 0)
binary_encoded = binary_encoded.astype(bool)

# Apply Apriori algorithm
frequent_itemsets = apriori(binary_encoded, min_support=0.2, use_colnames=True)

# Display frequent itemsets
print(frequent_itemsets)
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori


# Function to extract clean item names
def extract_clean_items(row):
    clean_items = []
    for item in row:
        if isinstance(item, str) and item != 'nan':
            clean_item = item.split('_')[0]
            if '(' in clean_item:
                clean_item = clean_item.split('(')[0].strip()
            clean_items.append(clean_item)
    return clean_items

# Apply the extract_clean_items function to create clean item columns
data['clean_items'] = data['item_purchased'].apply(extract_clean_items)

# Convert the 'clean_items' column to a list of lists
transaction_list = data['clean_items'].tolist()

# Convert the list of lists to a binary-encoded DataFrame
binary_encoded_df = pd.get_dummies(pd.DataFrame(transaction_list), prefix='', prefix_sep='')

# Apply Apriori algorithm
frequent_itemsets = apriori(binary_encoded_df, min_support=0.2, use_colnames=True)

# Display frequent itemsets
print(frequent_itemsets)
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder

# Assuming you have the clean item names in data['clean_items']
transaction_list = data['clean_items'].tolist()

# Initialize a TransactionEncoder
te = TransactionEncoder()

# Fit and transform the transaction list
te_ary = te.fit(transaction_list).transform(transaction_list)
#print(te_ary)
# Convert the encoded array into a DataFrame
encoded_df = pd.DataFrame(te_ary, columns=te.columns_)
#print(encoded_df)
# Apply the Apriori algorithm
frequent_itemsets = apriori(encoded_df, min_support=0.05, use_colnames=True)

# Display frequent itemsets
print(frequent_itemsets)
```

```python
import pandas as pd
import re
import matplotlib.pyplot as plt


# Extract item quantities using regular expressions
item_quantities = []
for items in data['item_purchased']:
    quantities = []
    for item in items:
        match_ml = re.search(r'\((\d+)ml\)', item)
        match_Ml = re.search(r'\[(\d+) Ml\]', item)
        match_quantity = re.search(r'_(\d+)_', item)
        
        if match_ml:
            quantities.append(int(match_ml.group(1)))
        elif match_Ml:
            quantities.append(int(match_Ml.group(1)))
        elif match_quantity:
            quantities.append(int(match_quantity.group(1)))
        else:
            quantities.append(1)  # Default to quantity 1 if no match is found
            
    item_quantities.extend(quantities)

# Create a frequency distribution
quantity_counts = pd.Series(item_quantities).value_counts()

# Calculate relative frequencies
total_obs = len(item_quantities)
relative_frequencies = quantity_counts / total_obs

# Visualize the distribution
plt.bar(relative_frequencies.index, relative_frequencies.values)
plt.xlabel('Item Quantities')
plt.ylabel('Relative Frequency')
plt.title('Distribution of Item Quantities Ordered')
plt.xlim(1, 10)  # Adjust x-axis limits
plt.show()

print(quantity_counts)
print(relative_frequencies)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Ensure 'Order-delivery-time' is in datetime format
data['Order-delivery-time'] = pd.to_datetime(data['Order-delivery-time'])

# Extract hour and day components
data['Hour'] = data['Order-delivery-time'].dt.hour
data['Day'] = data['Order-delivery-time'].dt.day_name()

# Group by hour and calculate order counts
hourly_order_counts = data.groupby('Hour')['Order-delivery-time'].count()

# Group by day and calculate order counts
daily_order_counts = data.groupby('Day')['Order-delivery-time'].count()

# Plotting
plt.figure(figsize=(12, 6))

# Plot for hourly order counts
plt.subplot(1, 2, 1)
hourly_order_counts.plot(kind='line', marker='o')
plt.xlabel('Hour')
plt.ylabel('Order Count')
plt.title('Hourly Order Frequency')

# Plot for daily order counts
plt.subplot(1, 2, 2)
daily_order_counts.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']).plot(kind='bar', color='skyblue')
plt.xlabel('Day')
plt.ylabel('Order Count')
plt.title('Daily Order Frequency')

#plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Assuming you have the clean item names in data['clean_items']
transaction_list = data['clean_items'].tolist()

# Initialize a TransactionEncoder
te = TransactionEncoder()

# Fit and transform the transaction list
te_ary = te.fit(transaction_list).transform(transaction_list)

# Convert the encoded array into a DataFrame
encoded_df = pd.DataFrame(te_ary, columns=te.columns_)

# Apply the Apriori algorithm
frequent_itemsets = apriori(encoded_df, min_support=0.01, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)

# Drop duplicates where antecedents and consequents are swapped
unique_rules = rules.drop_duplicates(subset=['antecedents', 'consequents'])

# Sort the rules by confidence in descending order
unique_rules = unique_rules.sort_values(by='confidence', ascending=False)

# Display the most frequently bought together itemsets (top 5)
unique_itemsets = unique_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(5)
print(unique_itemsets)
```

```python


# Define a function to extract the first part of each item
def extract_item_name(item):
    if isinstance(item, str):
        return item.split('_')[0]
    return None

# Apply the extract_item_name function to each element in the 'item_purchased' column
data['items'] = data['item_purchased'].apply(lambda items: [extract_item_name(item) for item in items if item != 'nan'])

print(data['items'])
```

```python
data['Order-relay-time(ordered time)'] = pd.to_datetime(data['Order-relay-time(ordered time)'], format='%d-%m-%Y %H:%M', errors='coerce')
```

```python
monthly_sales = data.groupby(pd.Grouper(key='Order-relay-time(ordered time)', freq='M'))['Total-bill-amount <bill>'].sum()
print(monthly_sales)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.facecolor'] = '#FFA726'

plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar',color='blue')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
#plt.grid(True)
plt.show()
```

```python
weekly_sales = data.groupby(pd.Grouper(key='Order-relay-time(ordered time)', freq='W'))['Total-bill-amount <bill>'].sum()
print(weekly_sales)
```

```python
# Function to extract item names from a list of items
def extract_item_names(item_list):
    extracted_items = []
    for item in item_list:
        if isinstance(item, str) and item != 'nan':
            extracted_item = item.split('_')[0].split()[0]
            extracted_items.append(extracted_item)
    return extracted_items

# Create the 'list_item' column by applying the extract_item_names function
data['list_item'] = data['item_purchased'].apply(extract_item_names)

print(data['list_item'])
```

```python
import re

def clean_item_name(item_name):
    # Remove content within parentheses or square brackets
    item_name = re.sub(r'\([^)]*\)', '', item_name)  # Removes content within parentheses
    item_name = re.sub(r'\[[^\]]*\]', '', item_name)  # Removes content within square brackets
    
    # Remove extra spaces
    item_name = ' '.join(item_name.split())
    
    # Remove digits, special characters, and trailing spaces
    item_name = re.sub(r'[0-9!@#$%^&*(),.?":{}|<>]', '', item_name)
    item_name = item_name.rstrip()
    
    return item_name

# Apply the clean_item_name function to the 'items' column
data['clean_items'] = data['list_item'].astype(str).apply(clean_item_name)
```
